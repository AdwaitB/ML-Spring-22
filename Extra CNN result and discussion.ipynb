// After the CNN 'Overall the accuracy' //

The Gradient-weight class activation mapping was used to see how the convolutional layers predicted the COVID positivity. 

The blue color shows that the convolutional layers focused on lung region of the X-ray to determine the positive COVID case. For negeative case, convolution highlighted non-lung region.

Next, the X-ray images have been masked with the lung region resulting only lung images in X-ray. 


<center>(Left original and Right masked)</center>

The CNN model was trained with the image on the right expecting that the performance would increase. The mask allows unncessary part of the image to be taken away.

However, the result shows that masking the images produced lower accuracy in predicting the COVID. The CNN architecture was kept the same and other hyperparameters were tested. The best result was obtained from changing the batch size to 32 where the original CNN model was trained with batch size of 256. It is interesting to notice that the performance decreased from 95% to 91%. One possible reason for lower accuracy can be concluded from the GRAD cam mapping. The CNN layers focused on the non-lung region to predict negative COVID. The mask image does not have other body parts except the lung region. Therefore, CNN model could not perform well on the predicting negative case.
